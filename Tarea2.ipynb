## Conjunto de Datos

Para este proyecto estare utilizando el dataset [Video Game Sales with Ratings](https://www.kaggle.com/datasets/rush4ratio/video-game-sales-with-ratings) que se encuentra en Kaggle.

Este dataset contiene una lista de video juegos junto con sus años y plataformas de lanzamiento, su genero, su desarrollador y publisher, sus ventas totales y por region, y las calificaciones tanto de los medios especializados como de los usuarios.

```python
df = pd.read_csv("Video_Games_Sales_as_at_22_Dec_2016.csv")
df.head(10)
```

## Modificación de Columnas

Añado 4 columnas () para describir la proporción de ventas por cada region. Además en base al rating se creo la columna Appropiate_Ages para definir la edad minima para jugar el juego.

```python
df_clean = df.dropna().copy()

df_clean['NA_Sales_Prop'] = df_clean['NA_Sales'] / df_clean['Global_Sales']
df_clean['EU_Sales_Prop'] = df_clean['EU_Sales'] / df_clean['Global_Sales']
df_clean['JP_Sales_Prop'] = df_clean['JP_Sales'] / df_clean['Global_Sales']
df_clean['Other_Sales_Prop'] = df_clean['Other_Sales'] / df_clean['Global_Sales']

esrb_ages = {
    'E': 6,
    'M': 17,
    'T': 13,
    'E10+': 10,
    'K-A': 6,
    'AO': 18,
    'EC': 3,
    'RP': pd.NA,
}
df_clean['Appropiate_Ages'] = df_clean['Rating'].map(esrb_ages)
df_clean = df_clean.dropna()

df_clean.head(10)
```

## Filtrar Datos

Filtramos los juegos que solo tengan ventas mayores que la media.

```python
over_mean_sales = df_clean[df_clean['Global_Sales'] > df_clean['Global_Sales'].mean()]
over_mean_sales
```

## Subconjuntos de Datos

Creamos un subconjunto con los juegos desarrollados por nintendo y sus ventas totales.

```python
nintendo_developed_games = df_clean.loc[df_clean['Developer'] == 'Nintendo', ['Name', 'Platform', 'Year_of_Release', 'Developer', 'Genre', 'Global_Sales']]
nintendo_developed_games
```

## Grupo de Datos

Se agrupan los datos por su rating, y se obtienen tanto las ventas promedios como las totales de cada rating.

```python
df_clean.groupby('Rating')['Global_Sales'].mean()
```

```python
df_clean.groupby('Rating')['Global_Sales'].sum()
```
