Aquí dejo un ejemplo de la seria fibonacci recursiva!

```python
def fibonacci(n: int) -> int:
    """
    Obtener el número de la posición n de la secuencia fibonacci.
    :param n: La posición del numero deseado.
    :returns: El número fibonacci de la posición indicada.
    """
    if not isinstance(n, int):
        raise TypeError
    if n <= 0:
        return 0
    if n == 1:
        return 1
    return fibonacci(n - 2) + fibonacci(n - 1)
```

```python
fibonacci(1)
```

```python
fibonacci(2)
```

```python
fibonacci(14)
```

```python
for i in range(1, 8):
    print(f'fibonacci({i}): {fibonacci(i)}')
```
